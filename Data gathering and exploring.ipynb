```python
import pandas as pd

# Load dataset
file_path = "/mnt/data/Raw-Data.csv"
df = pd.read_csv(file_path)

# Step 1: Handle Missing Values
# Fill missing categorical values with 'Unknown'
categorical_cols = ['Country', 'Age', 'Gender', 'Experiencing_Symptoms', 'Severity', 'Contact']
df[categorical_cols] = df[categorical_cols].fillna('Unknown')

# Step 2: Standardize Categories
# Replace 'Dont-Know' with 'Unknown' in the 'Contact' column
df['Contact'] = df['Contact'].replace('Dont-Know', 'Unknown')

# Step 3: Convert Age Ranges into Categories
def categorize_age(age):
    if age in ['0-9', '10-19']: return 'Child'
    elif age in ['20-24', '25-59']: return 'Adult'
    elif age == '60+': return 'Senior'
    else: return 'Unknown'

df['Age_Group'] = df['Age'].apply(categorize_age)

# Step 4: Remove duplicates
df = df.drop_duplicates()

# Step 5: Save cleaned data
cleaned_file_path = "/mnt/data/Cleaned-Data.csv"
df.to_csv(cleaned_file_path, index=False)

print("Data cleaning completed. Cleaned file saved at:", cleaned_file_path)
```

```python
import pandas as pd

# Load the dataset
file_path = "/mnt/data/Raw-Data.csv"
df = pd.read_csv(file_path)

# Step 1: Check for missing values
print("Missing values before cleaning:\n", df.isnull().sum())

# Step 2: Replace missing values with the mode of each column
for column in df.columns:
    mode_value = df[column].mode()[0]  # Get the mode of the column
    df[column].fillna(mode_value, inplace=True)  # Replace NaN with mode

# Step 3: Verify that there are no missing values left
print("Missing values after cleaning:\n", df.isnull().sum())

# Save the cleaned data
cleaned_file_path = "/mnt/data/Cleaned-Data.csv"
df.to_csv(cleaned_file_path, index=False)

print("Data cleaning completed. Cleaned file saved as 'Cleaned-Data.csv'.")
```
